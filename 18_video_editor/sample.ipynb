```python
import matplotlib.pyplot as plt
import numpy as np
from moviepy.editor import VideoClip
from moviepy.video.io.bindings import mplfig_to_npimage
```

```python
x = np.linspace(-2, 2, 200)

duration = 2

fig, ax = plt.subplots()
```

```python
def make_frame(t):
    ax.clear()
    ax.plot(x, np.sinc(x**2) + np.sin(x + 2*np.pi/duration * t), lw=3)
    ax.set_ylim(-1.5, 2.5)
    return mplfig_to_npimage(fig)
```

```python
animation = VideoClip(make_frame, duration=duration)
animation.ipython_display(fps=20, loop=True, autoplay=True)
```

```python
# code to write the animation to a file

animation.write_videofile("sinc.mp4", fps=20)
```

```python
# to add animation to an image file (gif, png, jpg, etc) use the following code

from moviepy.editor import ImageClip

image = ImageClip("myimage1.jpg")
image = image.set_duration(duration)
animation = animation.set_start(image.duration)
final_clip = ImageClip([image, animation])
final_clip.write_videofile("myimage1_with_animation.mp4", fps=20)

```
